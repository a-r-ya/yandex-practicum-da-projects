# Исследование факторов успешности компьютерных игр

**Описание проекта**

Цель проекта — выявить определяющие успешность игры закономерности для интернет-магазина компьютерных игр. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Описание данных**

Исторические данные из открытых источников до 2016 года о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation).

> - `Name` — название игры
> - `Platform` — платформа
> - `Year_of_Release` — год выпуска
> - `Genre` — жанр игры
> - `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
> - `EU_sales` — продажи в Европе (миллионы проданных копий)
> - `JP_sales` — продажи в Японии (миллионы проданных копий)
> - `Other_sales` — продажи в других странах (миллионы проданных копий)
> - `Critic_Score` — оценка критиков (максимум 100)
> - `User_Score` — оценка пользователей (максимум 10)
> - `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Предобработка данных

В этом разделе мы подготовим данные для дальнейшего анализа: приведём названия столбцов к нижнему регистру, обработаем пропущенные значения, заменим типы данных на корректные, добавим новые столбцы.

### Приведение названий столбцов к нижнему регистру

С помощью цикла создадим словарь с заменами старых названий столбцов на новые в нижнем регистре. Передадим этот словарь методу `rename`.

```python
columns_dict = {}                           # Создаем пустой словарь

for col in df.columns:                      # Заполняем словарь ключами (старое название) и значениями (новое)
    columns_dict[col] = col.lower()
    
df = df.rename(columns = columns_dict)      # Передаем словарь методу rename, перезаписываем датафрейм
```

### Обработка пропущенных значений

Посчитаем количество и процент пропущенных значений для каждого столбца.

```python
# Количество пропусков по столбцам с градиентной заливкой

pd.DataFrame({'n': df.isna().sum(),
              '%': df.isna().mean()})\
.style.background_gradient('coolwarm')
```

Пропуски найдены в 6 столбцах:
- `name` (название игры);
- `year_of_release` (год выпуска);
- `genre` (жанр игры);
- `critic_score` (оценка критиков);
- `user_score` (оценка пользователей);
- `rating` (рейтинг от организации ESRB).

3 из 6 столбцов содержат > 40% пропусков.

Возможные причины появления пропусков:
- в названии, годе выпуска, жанре игры — техническая ошибка или неполнота данных в открытых источниках;
- в оценках критиков и пользователей — игра старая и в год её выхода не было ресурсов, собирающих оценки, или игра малоизвестная, в связи с чем не привлекла должного внимания критиков;
- в рейтинге от ESRB — в год выхода игры ещё не присуждался возрастной рейтинг.

Для пропусков в каждом столбце будет свой алгоритм действий:

* В `name` и `genre` **удалим пропуски**, так как их очень мало (2) и без этой информации не получится провести анализ. Более того, пропуски взаимосвязаны и данные по двум столбцам отсуствуют в одних и тех же некачественных строках. 

```python
# Выведем строки с пропусками в названии или жанре

# df.query('name.isna() | genre.isna()') - старый код
```

```python
df.loc[(df['name'].isna()) | (df['genre']).isna()]
```

```python
# Удаление пропусков в столбцах name и genre

df = df.dropna(subset=['name', 'genre'])
```

* В `year_of_release`, `critic_score` и `user_score` **заменим пропуски на -1**. Позже мы поменяем формат на числовой и конвертация не получится, если предварительно не заменить `nan` числом. Год и оценки нельзя заменить медианами и другими синтетическими данными, так как это исказит результат, поэтому оставим пропуски как есть, поставив индикаторы для возможности отфильтровать такие пропуски при анализе. Мы не используем 0, чтобы не спутать пропуски с нулевыми оценками пользователей и критиков.

Столбцы `year_of_release` и `critic_score` имеют формат `float64`, поэтому единственно возможный вид пропусков в них `nan`, который мы можем сразу заменить.

```python
# Замена пропусков в year_of_release и critic_score

columns_to_fill = ['year_of_release', 'critic_score']
df[columns_to_fill] = df[columns_to_fill].fillna(-1)
```

Столбец `user_score` был в формате `object` вместо `float64` и вид пропусков в нём вызывает вопросы. Посмотрим на уникальные значения.

```python
# Смотрим на уникальные значения в user_score

df['user_score'].unique()
```

Помимо `nan` обнаруживаем значение `tbd`. [На сайте Metacritic указано](https://www.metacritic.com/faq#item13), что рейтинг "tbd" (англ. *to be determined*, "предстоит определить") присваивается, если оценок меньше 4. Получается, что для наших данных это ещё одна разновидность пропусков, которую мы также заменим.

```python
# Замена tbd на -1

df.loc[df['user_score'] == 'tbd', 'user_score'] = -1
```

```python
# Замена nan на -1

df['user_score'] = df['user_score'].fillna(-1)
```

* В `rating` **заменим пропуски на missing**, также добавив индикатор, потому что нет возможности заполнить данные.

```python
# Замена nan на -1

df['rating'] = df['rating'].fillna('missing')
```

### Изменение типов данных

Изменим типы данных:
- `year_of_release` переведём из `float64` в `int64`;
- `critic_score` переведём из `float64` в `int64`;
- `user_score` переведём из `object` в `float64`.

```python
# Замена чисел с плавающей точкой на целые числа

columns_to_convert = ['year_of_release', 'critic_score']
df[columns_to_convert] = df[columns_to_convert].astype('int64')
```

```python
# Замена объекта на число с плавающей точкой

df['user_score'] = df['user_score'].astype('float64')
```

```python
# Проверка: теперь все ок

df.info()
```

### Создание нового столбца с суммарными продажами

Создадим новый столбец для удобства анализа: суммарные продажи для игры во всех регионах `total_sales`.

#### Сколько игр выпускалось в разные годы?

```python
# Распределение выпущенных игр по годам (без учета пропусков)

sns.countplot(
    x=df[df['year_of_release'] != -1]['year_of_release'], # Исключение пропусков по годам (-1)
    palette='Blues'
)    
plt.title('Распределение игр по годам') 
plt.xlabel('год')
plt.ylabel('количество')
plt.xticks(rotation=90)
```

С тех пор последние 5 лет на графике выглядят стабильно: новых игр около 500-600 в год и этот период мы будем считать наиболее актуальным.

#### Как влияют на продажи отзывы пользователей и критиков?

Проверим, есть ли линейная зависимость между оценками пользователей / критиков и продажами игр внутри самой продаваемой в 2016 году платформы PS4.

```python
plt.figure(figsize=(12.8, 5)) 
plt.suptitle('Диаграммы рассеяния по платформе PS4, 2012 — 2016')

plt.subplot(1, 2, 1)
sns.scatterplot(
    data=df_relevant[(df_relevant['user_score'] != -1) & (df_relevant['platform'] == 'PS4')],
    x='user_score',
    y='total_sales'
)
plt.title('Рейтинг пользователей и глобальные продажи')
plt.xlabel('рейтинг пользователей')
plt.ylabel('продажи, млн копий')

plt.subplot(1, 2, 2)
sns.scatterplot(
    data=df_relevant[(df_relevant['critic_score'] != -1) & (df_relevant['platform'] == 'PS4')],
    x='critic_score',
    y='total_sales'
)
plt.title('Рейтинг критиков и глобальные продажи')
plt.xlabel('рейтинг критиков')
plt.ylabel('продажи, млн копий')
```

```python
# Коэффициент Пирсона (без учета пропусков -1)

for col in ['user_score', 'critic_score']:
    r = df_relevant[(df_relevant[col] != -1) & (df_relevant['platform'] == 'PS4')][col].corr(
        df_relevant[(df_relevant[col] != -1) & (df_relevant['platform'] == 'PS4')]['total_sales'])
    print(f'r для total_sales и {col}: {round(r, 2)}')
```

Коэффициент корреляции Пирсона `-0.03` показывает отсутствие связи между оценкой пользователей и продажами, а `0.41` — умеренную положительную связь между оценкой критиков и продажами.

Среди игр с низкими пользовательскими рейтингами встречаются выбросы с высокими продажами. Возможно, что некоторые игры хорошо рекламировались и покупались, но впоследствии не оправдали ожиданий игроков. Также есть обратные случаи, когда высокие оценки игроков получили игры с невысокими продажами. Мы не можем говорить о наличии связи из-за таких случаев. Игроки могут не оценить популярные игры, но полюбить что-то более нишевое.

Однако между оценками критиков и продажами наблюдается корреляция, которая видна на графике. Либо благосклонность критиков заставляет покупать игры, либо критики сходятся во мнении с игроками, проголосовавшими кошельком, и в топе мы видим действительно качественные продукты, которые одновременно высоко оценены и популярны (корреляция ничего не говорит о причинно-следственных связях, это только предположения).

Вычислим коэффициенты корреляции для других топовых платформ 2016 года.

```python
# Коэффициент Пирсона по платформам

for pf in pivot_top_2016['platform']:
    if pf != 'PS4':
        for col in ['user_score', 'critic_score']:
            r = df_relevant[(df_relevant[col] != -1) & (df_relevant['platform'] == pf)][col].corr(
                df_relevant[(df_relevant[col] != -1) & (df_relevant['platform'] == pf)]['total_sales'])
            print(f'{pf}: r для total_sales и {col}: {round(r, 2)}')
        print()
```

В целом, ситуация с другими платформами похожа PS4. Почти отсутствует корреляция продаж с оценками пользователей, но есть умеренная корреляция продаж с оценками критиков. Исключение — WiiU: умеренная корреляция продаж с оценками пользователей.

## Проверка гипотез

В этом разделе мы проведём проверку гипотез с помощью статистических тестов. Нас интересует, есть ли статистически значимые отличия между пользовательскими рейтингами разных платформ и жанров внутри актуального периода.

### Проверка гипотезы: средние пользовательские рейтинги платформ Xbox One и PC различаются

- H_0: Средний пользовательский рейтинг (user_score) платформы Xbox One = средний пользовательский рейтинг (user_score) платформы PC
- H_a: Средний пользовательский рейтинг (user_score) платформы Xbox One ≠ средний пользовательский рейтинг (user_score) платформы PC
- $\alpha$ = 0.05

Мы сформулировали нулевую гипотезу о равенстве средних двух генеральных совокупностей. Нас интересует отклонение как в большую, так и в меньшую сторону, поэтому альтернативная гипотеза о неравенстве средних — двусторонняя. 

Пороговое значение $\alpha$ установили на уровне 0.05: мы принимаем 5% вероятность получить наблюдаемый результат при условии, что нулевая гипотеза верна, как достаточно малую, и на этом основании будем отвергать нулевую гипотезу.

Так как мы будем сравнивать **средние двух генеральных совокупностей между собой**, мы выбираем метод **t-теста** (t-критерий Стьюдента). Убедимся, что для теста выполнены основные допущения (assumptions):
- Генеральные совокупности независимы? ✅ Нет оснований полагать, что зависимы.
- Выборочные средние распределены нормально? ✅ Выборки достаточно большие -> ЦПТ.
- Дисперсии генеральных совокупностей равны? ✅ Выборки достаточно большие; нет оснований полагать, что дисперсии не равны.

```python
# Функиция для t-теста для 2 независимых выборок с визуализацией распредлений и тестом Levene

def t_test(sample_1, sample_2, alpha):
    try:
        fig, (ax1, ax2) = plt.subplots(2)                                    # Гистограммы
        fig.suptitle('Распределения выборок')
        sample_1.hist(ax=ax1)
        sample_2.hist(ax=ax2)
        
        levene = st.levene(sample_1, sample_2)                               # Дисперсии равны?
        print('Levene test p-value:', levene.pvalue)
        
        if levene.pvalue < alpha:                                              
            print('Отвергаем нулевую гипотезу о равенстве дисперсий')
            print()
            print("Проведён Welch's t-test")
            results = st.ttest_ind(sample_1, sample_2, equal_var=False)      # t-test equal_var=False
        else:
            print('Не получилось отвергнуть нулевую гипотезу о равенстве дисперсий')
            print()
            print("Проведён Student's t-test")
            results = st.ttest_ind(sample_1, sample_2, equal_var=True)       # t-test equal_var=True
            
        print('T-test p-value:', results.pvalue)
        if results.pvalue < alpha:
            print('Отвергаем нулевую гипотезу')
        else:
            print('Не получилось отвергнуть нулевую гипотезу')
    except:
        print('Неверные данные')
```

```python
# Проводим сам тест. Не забудем исключить из анализа пропуски (-1)

xone_score = df_relevant[(df_relevant['user_score'] != -1) & (df_relevant['platform'] == 'XOne')]['user_score']
pc_score = df_relevant[(df_relevant['user_score'] != -1) & (df_relevant['platform'] == 'PC')]['user_score']

t_test(xone_score, pc_score, 0.05)
```

Вероятность получить наблюдемый результат при условии, что H_0 верна, составляет 5%. На данном уровне значимости нет оснований отвергать H_0. Средний пользовательский рейтинг платформы Xbox One **статистически значимо не отличается** от среднего пользовательского рейтинга платформы PC.

### Проверка гипотезы: средние пользовательские рейтинги жанров Action и Sports различаются

- H_0: Средний пользовательский рейтинг (user_score) жанра Action = средний пользовательский рейтинг (user_score) жанра Sports
- H_a: Средний пользовательский рейтинг (user_score) жанра Action ≠ средний пользовательский рейтинг (user_score) жанра Sports
- $\alpha$ = 0.05

Проверяем аналогично предыдущей гипотезе.
