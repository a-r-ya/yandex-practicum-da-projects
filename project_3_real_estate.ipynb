### Обработка пропущенных значений

Посчитаем количество пропущенных значений для каждого столбца.

```python
# Количество пропусков по столбцам

df.isna().sum()
```

Пропуски найдены в 14 столбцах. Посмотрим, какую долю составляют пропуски от общего числа наблюдений.

```python
# Хитмэп, показывающий пропуски по столбцам. Данные транспонировали, чтобы по оси y были столбцы, убрали легенду и подпись оси x

sns.heatmap(
    df.isna().transpose(), 
    cbar=False, 
    xticklabels=False, 
    cmap='Pastel1_r'
)

plt.title('Матрица пропущенных значений по столбцам')
```

Видим, что данные довольно "сырые", в них много пропусков, особенно в столбце `is_apartment`. Построим более наглядную визуализацию, где отсортируем столбцы по доле пропусков.

```python
# Столбчатая диаграмма с накоплением для пропусков + полных данных

(
    df.isna()                                          # Получаем массив из наших данных с булевыми значениями пропусков
    .melt()                                            # Преобразуем столбцы в строки
    .pivot_table(                                      # По этим строкам строим сводную, где указываем
        index='variable',                              # индекс (столбец исходного датафрейма),
        aggfunc=['sum', lambda x: len(x) - sum(x)]     # вычисляемые столбцы (сумма всех True, т.е. пропусков, и сумма всех
    )                                                  # False, т.е. не-пропусков через лямбда-функцию)
    .reset_index()                                     # Сбрасываем индекс, чтобы он записался в отдельный столбец
    .sort_values(by=('sum', 'value'))                  # Сортируем по сумме пропусков для каждого столбца
    .plot(                                             # Строим график
        'variable',                                    
        kind='barh', 
        stacked=True, 
        legend=False, 
        cmap='Pastel1', 
        title='Доля пропущенных значений по столбцам'
    )
)

plt.ylabel('')                                         # Убираем подписи к оси y

treshold = df.shape[0] / 100 * 5                       # Добавляем линию с 5% порогом
plt.axvline(                                           # Преположим, что до порога можно отбросить пропуски
    x=treshold,                         
    color='red', 
    ls=':',
)
plt.text(treshold, 0, '5%', color='red')
```

12 из 14 столбцов содержат > 5% пропусков, что не очень хорошо.

Возможные причины появления пропусков:
- в добавленных пользователем данных — пользователь забыл указать все параметры и система не требует их обязательного заполнения;
- в картографических данных — проставляя геоданные, система не указывает нули, если не нашла парков / водоёмов / аэропортов поблизости;
- в `days_exposition` — если квартира ещё не продана, для неё не высчитыватся количество дней до продажи.

Для пропусков в каждом столбце будет свой алгоритм действий:

* В `ceiling_height` **заменим пропуски медианным значением в зависимости от расстояния до центра**. В данных нет столбца с типом дома, но можно косвенно понять тип застройки по расстоянию до центра Санкт-Петербурга в `cityCenters_nearest`. Мы предполагаем, что в центре сосредоточится старый фонд с высокими потолками, для более удалённых районов и Ленинградской области будет характерна типовая застройка с низкими потолками. 

```python
# Проверим гипотезу о разных потолках для разной удаленности от центра, посмотрев на медиану: действительно, есть такой паттерн

df.groupby('cityCenters_nearest')['ceiling_height'].median()
```

```python
# Замена пропусков в столбце ceiling_height с высотой потолков медианой по cityCenters_nearest с расстоянием до центра

citycenter_median = df.groupby('cityCenters_nearest')['ceiling_height'].transform('median')
df['ceiling_height'] = df['ceiling_height'].fillna(citycenter_median)
```

```python
# Так как не по каждому расстоянию могут быть данные, заменим оставшиеся пропуски высотой для ближайшего расстояния

df = df.sort_values(by='cityCenters_nearest')
df['ceiling_height'] = df['ceiling_height'].ffill()
```

* В `floors_total` **удалим пропуски**, так как их всего 86 (менее 5%) и без данных об этажности дома невозможно провести анализ.

```python
# Удаление пропусков в столбце floors_total с этажностью дома

df = df.dropna(subset=['floors_total'])
```

* В `living_area` и `kitchen_area` **оставим пропуски**, так как заполняя площади синтетическими данными, можно исказить результат. Можно было бы предположить, что пропуски характерны только для квартир-студий, где нет чёткого разделения на жилую и кухонную площадь, но в наших данных пропуски несистематичны. 

```python
df.loc[df['studio'], ['living_area', 'kitchen_area']].isna().sum()
```

```python
df.loc[~df['studio'], ['living_area', 'kitchen_area']].isna().sum()
```

* В `is_apartment` **заменим пропуски на `False`**, так как данные должны быть булева типа и мы предполагаем, что все апартаменты уже имеют соответствующую пометку.

```python
# Заполнение пропусков значением False в данных булева типа is_apartment с апартаментами

df['is_apartment'] = df['is_apartment'].fillna(False)
```

```python
# После заполнения пропусков is_apartment должен стать bool вместо object и нам уже не нужно будет менять тип - проверим

df['is_apartment'].dtypes
```

* В `balcony`, `parks_around3000` и `ponds_around3000` **заменим пропуски на `0`**, так как предполагаем, что балкон является преимуществом при продаже квартиры и был бы указан, следовательно, его нет; данные по ближайшим паркам и водоёмам автоматически проставляются системой, следовательно, их тоже нет.

```python
# Заполнение пропусков нулями в столбцах с точным количеством - балконов, парков и водоёмов поблизости

columns_to_fill = ['balcony', 'parks_around3000', 'ponds_around3000']
df[columns_to_fill] = df[columns_to_fill].fillna(0)
```

* В `locality_name` **заменим пропуски на "Санкт-Петербург"**, если расстояние до центра Санкт-Петербурга `cityCenters_nearest` меньше или равно 15 км, так как это примерное расстояние до окраин. В остальных случаях **удалим пропуски**, так как всего их 49 (менее 5%) и без данных о населённом пункте невозможно провести анализ.

```python
# Боксплот для расстояния от центра по объявлениям в Санкт-Петербурге. Основная часть данных действительно лежит в диапазоне 8-15 км
 
df[df['locality_name'] == 'Санкт-Петербург'].boxplot(
    column='cityCenters_nearest',
    figsize=(4, 4)
)
plt.title('Диаграмма размаха для расстояния до центра среди квартир в Санкт-Петербурге')
```

```python
# Замена пропусков в столбце locality_name с названием населенного пункта на "Санкт-Петербург", если до центра <= 15 км

df.loc[(df['locality_name'].isna()) & (df['cityCenters_nearest'] <= 15000), 'locality_name'] = 'Санкт-Петербург'
```

Причиной дубликатов становятся разница в написании "е"/"ё" и отсутствие категоризации населенного пункта. 

Чтобы убрать дубликаты, заменим все "ё" на "е" и с помощью пользовательской функции получим названия населенных пунктов без категоризации.

```python
# Замена "ё" на "е" в столбце locality_name с названиями населенного пункта

df['locality_name'] = df['locality_name'].str.replace('ё', 'е')
```

```python
'''
Функция для получения краткого названия населенного пункта - только имя собственное.
На вход подается строка с названием населенного пункта через пробелы.
Из слов в строке формируется список.
Для каждого слова в списке проверяется, пишется ли оно с заглавной буквы.
Если условие выполнено, проверка прекращается, и в новый список записываются имена собственные, начиная с первого найденного.
Этот список преобразуется обратно в строку, где пробелами разделены полученные имена собственные.
Полученную строку и возвращает обратно функция.

* НА БУДУЩЕЕ: можно сделать элегантнее с регулярным выражением, надо изучить
'''

def get_locality_short(name):
    try:
        wordlist = name.split()
        for word in wordlist:
            if word.lower() != word:
                locality = wordlist[wordlist.index(word):]
                break
        locality = ' '.join(locality)
        return locality
    except:
        return None
```

```python
# Замена полных названий населенных пунктов в столбце locality_name на короткие с помощью пользовательской функции

df['locality_name'] = df['locality_name'].apply(get_locality_short)
```

```python
# Спискок уникальных заначений в столбце locality_name с названиями населенного пункта после обработки дубликатов

df['locality_name'].unique().tolist()
```

Создадим новые столбцы для удобства анализа:

- день публикации объявления `day_of_publication`;

- год публикации объявления `year_of_publication`;
- категория этажа квартиры (первый, последний, другой) `floor_category`;
- категория локации (Санкт-Петербург или пригород) `location_category`;

```python
# Создание нового столбца - год публикации объявления

df['year_of_publication'] = df['first_day_exposition'].dt.year
```

```python
# Создание нового столбца - тип этажа квартиры (значения — «первый», «последний», «другой»)

def get_floor_category(df):
    try:
        if df['floor'] == 1:
            return 'первый'
        elif df['floor'] == df['floors_total']:
            return 'последний'
        else:
            return 'другой'
    except:
        return None
    
df['floor_category'] = df.apply(get_floor_category, axis=1)

df['floor_category'].unique()
```

```python
# Создание нового столбца - расстояние в км до центра города (целочисленное)

df['km_from_the_centre'] = round(df['cityCenters_nearest'] / 1000)
```

### Как быстро продавались квартиры?

Теперь изучим, как быстро продавались квартиры. Какие продажи можно считать быстрыми, а какие — необычно долгими?

```python
# Гистограмма распределения + боксплот для продаж

plt.figure(figsize=(12.8, 5)) 
plt.suptitle('Распределение сроков продажи квартир')

plt.subplot(1, 2, 1)
sns.histplot(
    df,
    x='days_exposition',
    kde=True
)
plt.xlabel('срок продажи, дни')
plt.ylabel('количество')

plt.subplot(1, 2, 2)                                     
sns.boxplot(                            
    data=df,
    y='days_exposition'
)
plt.xlabel(None)
plt.ylabel('срок продажи, дни')
```

На графиках видим много выбросов, продававшихся больше 500 дней. Может быть, объявления забывали закрыть, если продажа становилась неактуальной? Построим гистограмму для продаж до 500 дней.

```python
# Гистограмма распределения для продаж (до 500 дней)

sns.histplot(
    df,
    x='days_exposition',
    kde=True,
    bins=300
)
plt.title('Распределение сроков продажи квартир до 500 дней')
plt.xlabel('срок продажи, дни')
plt.ylabel('количество')
plt.xlim(0, 500)
```

На гистограмме выбиваются два значения до 100 дней, что кажется необычным для такого вида распределения. Выведем эти наиболее частотные значения.

```python
# Выводим частоты для каждого срока продажи и сортируем по убыванию

df['days_exposition'].value_counts().sort_values(ascending=False)
```

В топе по частотности видим подозрительные значения: 45, 60, 7, 30 и 90. Квартира может продаваться полтора месяца, но настолько ли часто продажа занимает ровно 45 календарных дней? [Правила сервиса Яндекс.Недвижимость](https://yandex.ru/support/realty/owner/home/add-ads-housing.html) перечисляют стандартные сроки публикации объявлений без продления в зависимости от типа объекта, и мы можем обнаружить, что они совпадают с нашими странными значениями. Получается, что объявление могли выложить и не продлить его размещение, поэтому оно было закрыто системой через 7 / 30 / 45 / 60 / 90 дней. Это не обязательно значит, что произошла продажа.

```python
df[['days_exposition']].describe()
```

Медианное значение дней от публикации объявления до продажи составляет 96 дней. Основаная часть квартир (Q1-Q3) продалась за 45-232 дня, среднее значение составило 181 день, или примерно 6 месяцев.

Глядя на распределение, будем считать быстрой продажей ту, которая находится ниже первого квартиля, то есть меньше 45 дней. Добавим информацию о категории продажи в новый столбец `quick_sale`.

```python
# Создание нового столбца - быстрая продажа

df['quick_sale'] = df['days_exposition'].apply(
    lambda x: 'быстрая продажа' if x < 45 else 'обычная продажа'
)

df['quick_sale'].unique()
```

```python
# Создадим список из 10 населённых пунктов с наибольшим числом объявлений

top_localities = df['locality_name'].value_counts().head(10).index

top_localities
```

### Как средняя цена м² в Санкт-Петербурге зависит от расстояния до центра?

Вычислим среднюю цену за м² для каждого километра от центра Санкт-Петербурга. Для этого возмём срез объявлений только по Санкт-Петербургу и построим сводную таблицу, где каждому километру удалённости от центра будет соответствовать среднее значение цены за м².

```python
# Создаем сводную таблицу для визуализации средней цены каждого км в СПб

pivot_km_price = (                                                 # Сводная таблица для данных только по СПб
    df[df['locality_name'] == 'Санкт-Петербург']                     # по умолчанию для values посчитается среднее
    .pivot_table(
        index='km_from_the_centre', 
        values='price_per_sq_m'
    )
    .reset_index()
)

pivot_km_price
```

```python
# Скаттерплот для двух количественных переменных: цены за м2 и удаленности от центра в км

(
    sns.scatterplot(
        data=pivot_km_price, 
        x='km_from_the_centre', 
        y='price_per_sq_m'
    )
    .ticklabel_format(style='plain')                                               # Убрать e по оси y
)
plt.title('Диаграмма рассеяния для расстояния до центра Санкт-Петербурга и средней цены м²') 
plt.xlabel('расстояние до центра, км')
plt.ylabel('средняя цена м², руб.')
```

Центр Санкт-Петербурга действительно считается элитным местом: при расстоянии до 1 км средняя цена м² достигает значения более 225 тысяч рублей. Наиболее высокие средние цены м² встречаются до 5-7 км, после на графике виден драматический спад: по ценам мы можем наблюдать, где примерно заканчивается центр. После 10 км средняя цена м² за редкими исключениями не превышает 120 тысяч рублей и выглядит на графике однородной.

```python
# Коэффициент Пирсона

pivot_km_price['km_from_the_centre'].corr(pivot_km_price['price_per_sq_m'])
```

Коэффициент корреляции Пирсона `-0.78` показывает высокую отрицательную связь между удалённостью от центра и средней ценой м²: чем больше удалённость, тем меньше цена.
